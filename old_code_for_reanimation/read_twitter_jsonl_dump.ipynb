```python
# Read in just as much as I need from all the files
import pandas as pd
import json_lines
import json
import glob

keepcols = ['source']
datadir = 'kenya/'
maxfiles = -1

df = pd.DataFrame([])
for filename in glob.glob('{}*.jsonl'.format(datadir))[:maxfiles]:
    print(filename)
    items = []
    with open(filename, 'rb') as f:
        for item in json_lines.reader(f):
            items += [item]
    dffile = pd.DataFrame(items)
    dffile = dffile[keepcols]
    df = df.append(dffile, ignore_index=True)

df
```

```python
# Count up the sources in the dataset
dfsources = df['source'].value_counts().reset_index()
dfsources.columns = ['source', 'count']
dfsources['sourceurl'] = dfsources['source'].str.extract(r'\"(\S+)" ', expand=True)
dfsources['sourcetxt'] = dfsources['source'].str.extract(r'>(.*)<', expand=True)
outcsv = '{}_sources.csv'.format(datadir.strip('/'))
print('Writing source list to {}'.format(outcsv))
dfsources.to_csv(outcsv, index=False)
dfsources[['sourcetxt', 'sourceurl', 'count']]
```
